```python
import numpy as np
```

### `np.arange(start, stop, step)`
使用 `arange` 在 [start, stop] 区间内生成 step 为公差的等差数列。

```python
print(np.arange(1,100,2))
```

### `np.linspace(start, stop, num)`
使用 `linspace` 在 [start, stop] 区间内生成 num 个均匀分布的等差数列

参数：
- bool `endpoint`：是否包含 stop

```python
print(np.linspace(1, 100, 23))
```

```python
print(np.linspace(0,1,10))
print(np.linspace(-5,5,50))
print(np.linspace(10,20,5, endpoint=False))
print(np.linspace(0,250,100))
print(np.linspace(0,1,10,endpoint=False))
print(np.arange(0,1,0.1))
```
