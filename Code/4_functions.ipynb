### `np.where()` 二元运算

````python
import numpy as np
budgets = np.array([
    [34, 56, 23, 67, 45],
    [29, 61, 42, 38, 55],
    [47, 33, 60, 25, 41],
    [52, 68, 31, 44, 39],
    [58, 27, 50, 63, 36],
    [21, 48, 70, 30, 59],
    [65, 26, 37, 53, 46]
])
isHigh = np.where(budgets >= 60, '高', '低')
print(isHigh)
````

### `np.sort()` 排序

````python
import numpy as np
nums = np.array([1, 3, 2, 5])
numsSorted = np.sort(nums)
print(numsSorted)
````

### `np.ceil()` 向上取整

````python
import numpy as np
nums = np.array([4.3, 5.001, 6.7, 7.0])
numsCail = np.ceil(nums)
print(numsCail)
````

### `np.floor()` 向下取整

````python
import numpy as np
nums = np.array([4.3, 5.001, 6.7, 7.0])
numsFloor = np.floor(nums)
print(numsFloor)
````

````python
# 1. 分别使用
nums = np.array([3.2, 4.8, 5.5, 2.1])
numsFloor = np.floor(nums)
print(numsFloor)
numsCail = np.ceil(nums)
print(numsCail)
````

````python
# 2. 租车问题
print(np.ceil(43 / 10))
````

````python
# 3. 买书
print(np.floor(100 / 18))
````

def loadtxt(

    delimiter: str | None = None,
    converters: Mapping[int | str, (str) -> Any] | ((str) -> Any) | None = None,
    usecols: int | Sequence[int] | None = None,
    skiprows: int = 0,
